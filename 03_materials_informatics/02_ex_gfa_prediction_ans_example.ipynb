```python
import numpy as np
from sklearn.ensemble import RandomForestClassifier

from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score, recall_score, precision_score
from sklearn.model_selection import cross_val_score

from matminer.datasets import load_dataset
from matminer.featurizers.conversions import StrToComposition
from matminer.figrecipes.plot import PlotlyFig
```

Dataset description

glass_binary: Metallic glass formation data for binary alloys, collected from various experimental techniques such as melt-spinning or mechanical alloying. This dataset covers all compositions with an interval of 5 at. % in 59 binary systems, containing a total of 5959 alloys in the dataset. The target property of this dataset is the glass forming ability (GFA), i.e. whether the composition can form monolithic glass or not, which is either 1 for glass forming or 0 for non-full glass forming.


glass_binary_v2: Identical to glass_binary dataset, but with duplicate entries merged. If there was a disagreement in gfa when merging the class was defaulted to 1.

```python
df = load_dataset("glass_binary_v2")
```

```python
df.head(20)
```

```python
# Generate Composition from formula and add it to data frame
stc = StrToComposition()
df = stc.featurize_dataframe(df, "formula")
```

```python
df.head(10)
```

Descriptor generation

```python
# Create an object of a class that generates features of composition ratio
from matminer.featurizers.composition import ElementFraction
ef = ElementFraction()
```

```python
# Descriptor label
element_fraction_labels = ef.feature_labels()
print(element_fraction_labels)
```

```python
# Add descriptor
df = ef.featurize_dataframe(df, "composition")
```

```python
df.head(10)
```

Model training and prediction

```python
# Generate descriptor data matrix X, objective variable y
X = df.drop(["gfa", "formula", "composition"], axis=1)
y = df['gfa'].values
```

```python
# Prediction model (Random Forest) object generation, learning with training data
rf = RandomForestClassifier(n_estimators=100, random_state=1)
rf.fit(X, y)
```

```python
# Prediction for training data
y_pred = rf.predict(X)
```

```python
# Accuracy (predicted score)
acc = accuracy_score(y, y_pred)
print('Accuracy: ', acc)
```

```python
# Recall (predicted score)
rec = recall_score(y, y_pred)
print('Recall: ', rec)
```

```python
# Accuracy (predicted score)
pre = precision_score(y, y_pred)
print('Precision: ', pre)
```

Verification of test performance by cross-validation

```python
# Create object for cross-validation
kfold = KFold(n_splits=2, shuffle=True)
```

```python
# Accuracy of test data by cross-validation (predicted score)
scores = cross_val_score(rf, X, y, scoring='accuracy', cv=kfold)
print('Mean Accuracy: {:.3f}'.format(np.mean(scores)))
```

```python
# Recall (predicted score) of test data by cross-validation
scores = cross_val_score(rf, X, y, scoring='recall', cv=kfold)
print('Mean Recall: {:.3f}'.format(np.mean(scores)))
```

```python
# Accuracy of test data by cross-validation (predicted score) 
scores = cross_val_score(rf, X, y, scoring='precision', cv=kfold)
print('Mean Precision: {:.3f}'.format(np.mean(scores)))
```
